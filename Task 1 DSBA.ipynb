## Importing all the Required Modules


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression  
from sklearn.model_selection import train_test_split
from sklearn import metrics
import seaborn as sns
```

## Reading the data

```python
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
data = pd.read_csv(url)

#To check if data is read correctly
print("Data Successfully imported !")
data.head(10)
```

## Visualizing data - Histograms

```python
plt.subplots_adjust(wspace = 0.5)
plt.subplot(1, 2, 1)
data['Scores'].plot(kind = 'hist', color = 'red')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
data['Hours'].plot(kind = 'hist', color = 'blue')
plt.xlabel('Hours')
plt.ylabel('Frequency')
plt.locator_params(axis = 'y', nbins = 5)
plt.show()
```

# Plotting a graph to check the relation between scores and hours

```python
data.plot(x='Hours', y='Scores', style='s')
plt.title('Graph of Hours vs Scores of students')
plt.xlabel('Number of hours studied',size = 18)
plt.ylabel('Scores Achieved',size = 18)
plt.show()
```

### From the graph we can see, the relation between Hours and Scores is linear.

# Simple linear regression

```python
sns.set(rc = {'figure.figsize': (15, 7)}, font_scale = 1.2)
sns.regplot(x = 'Hours', y = 'Scores', data = data, scatter_kws = {'color': 'blue'}, color = 'green')
plt.title('Simple Linear Regression', size = 18)
plt.xlabel('Number of Hours studied', size = 15)
plt.ylabel('Score Achieved', size = 15)
plt.show()
```

### Simple linear regression is a statistical method that allows us to summarize and study relationships between two continuous (quantitative) variables:
#### One variable, denoted x, is regarded as the predictor, explanatory, or independent variable.
#### The other variable, denoted y, is regarded as the response, outcome, or dependent variable.

## Data Preprocessing

##### First we'll take the independant data (Hours) from the dataset and assign them to the variable X, and the dependant data (Scores) and assign them to the variable Y.



```python
#X denotes hours
#Y denotes score
X = data.iloc[:, :-1].values  
Y = data.iloc[:, 1].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
print("Training Set: ")
print("X:\n", X_train)
print("Y:\n", Y_train)
```

```python
print("Test Set: ")
print("X:\n", X_test)
print("Y:\n", Y_test)
```

## Training the Linear Regression Model

```python
regressor = LinearRegression()  
regressor.fit(X_train, Y_train)
print("Training complete Successfully.")
```

## Making Predictions

```python
Y_pred = regressor.predict(X_test)
```

## Predicting the test set scores using our Model

```python
df = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})  
df
```

```python
list1 = ['blue','yellow']
df.plot(kind = "bar",figsize=(10,10),color = list1)
plt.show()
```

```python
plt.scatter(X_test, Y_test, color = 'red', label = 'Actual Values')
plt.plot(X_test, Y_pred, color = 'green', label = 'Predicted Values')
plt.title('Comparison between the actual and predicted Value', size = 18)
plt.xlabel('Number of hours studied', size = 18)
plt.ylabel('Score Achieved', size = 18)
plt.legend()
plt.show()
```

## Evaluating the performance of the model

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))
 
```

## Predicting the score for a student who studied for 9.25 hours

```python
hours =[ [9.25] ]
own_pred = regressor.predict(hours)
plt.ylabel( 'Percentage Score', fontsize= "18")
plt.title('Predicted score if a student studies for 9.25 Hr/day' , fontsize="18")
plt.xlabel('No. of hours Studied' , fontsize ="18")
plt.plot(hours[0],own_pred [0],'go')
plt .show( )
print ("No of Hours studied = {} ".format(hours[0]))
print ("Predicted Score = "+str(own_pred[0])+" %")
```
